# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def _init_(self, marka,model,renk,kapasite):
        self.marka=marka
        self.model=model
        self.renk=renk
        self.kapasite=kapasite
    def seating_capacity(self,kapasite):
        return f"{self.marka}, {self.model} aracımızın kapasitesi {self.kapasite}"
class Bus (Vehicle) :
    def seating_capacity(self,kapasite = 50):
       self.kapasite=kapasite
       return super(). seating_capacity(kapasite=50)
schoolBus= Bus('mercedes','Benz','Siyah',32)
print(schoolBus.seating_capacity())

````
